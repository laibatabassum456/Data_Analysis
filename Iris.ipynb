1. Introduction

This project analyzes the Iris dataset to understand relationships between features and visualize data patterns.


2. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

3. Load Dataset

```python
df = sns.load_dataset('iris')
```

4. Explore Data:

```python
df.shape
df.columns
df.head()
df.describe()
```

5. Visualization

Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.show()
```

Histogram

```python
df['sepal_length'].hist()
plt.show()
```

Box Plot

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.show()
```

6. Conclusion
From the analysis, we observed clear differences between species based on petal length and sepal length.
